```python
import ast
import numpy as np
import pandas as pd
import os,sys
```

```python
t1Path = 'D:/mt_data/1126_withrw/encoded/t1/'
t2Path = 'D:/mt_data/1212_withrw/encoded/t1/'
```

```python
devices = ['pi3', 'pi4_2G', 'pi4_4G']
```

```python
features = [
    'system calls frequency_1gram',
    'system calls tfidf_1gram', 
    'system calls hashing_1gram',
    'system calls frequency_2gram', 
    'system calls tfidf_2gram',
    'system calls frequency_3gram',
    'system calls tfidf_3gram',
    'system calls frequency_2gram-pcas', 
    'system calls tfidf_2gram-pcas',
    'system calls frequency_3gram-pcas',
    'system calls tfidf_3gram-pcas', 
    'system calls frequency_4gram-pcas', 
    'system calls tfidf_4gram-pcas',
    'system calls frequency_5gram-pcas',
    'system calls tfidf_5gram-pcas',
    'system calls frequency_1gram-scaled', 
    'system calls tfidf_1gram-scaled',                
    'system calls frequency_2gram-scaled', 
    'system calls tfidf_2gram-scaled',
    'system calls frequency_3gram-scaled',
    'system calls tfidf_3gram-scaled', 
    ]
```

```python
def get_label(df):
    malware = df['maltype']
    label = []
    for m in malware:
        if m == 'normal':
            label.append(1)
        elif m == 'normalv2':
            label.append(2)
        else:
            label.append(-1)
    return label
```

```python
outputPath = 'D:/mt_data/data/'
```

```python
for f in features:
    device = []
    label = []
    malware = []
    data = []
    for d in devices:
        print(d,f)
        filename = 'encoded_bow{}_{}_{}.csv'.format(d, 60, f)
        t1 = t1Path+filename
        t1df = pd.read_csv(t1, sep='\t')
        t1data = [ast.literal_eval(i) for i in t1df[f]]
        t1device = [d+"_t1" for i in range(0, len(t1data))]
        t1label = get_label(t1df)

        t2 = t2Path+filename
        t2df = pd.read_csv(t2, sep='\t')
        if len(t2df)==361:
            t2df = t2df[0:-1]

        t2data = [ast.literal_eval(i) for i in t2df[f]]
        t2device = [d+"_t2" for i in range(0, len(t2data))]
        t2label = get_label(t2df)

        device += t1device
        device += t2device
        label += t1label
        label += t2label
        malware += t1df['maltype'].tolist()
        malware += t2df['maltype'].tolist()
        data += t1data
        data += t2data
    df = pd.DataFrame([device,label,malware,data]).transpose()
    df.columns=['device','label','malware','data']

    of = outputPath+'encoded_bow_{}.csv'.format(f)
    df.to_csv(of, sep='\t', index=None)
```
